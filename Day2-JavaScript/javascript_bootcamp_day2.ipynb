# JavaScript
### 1.1.  Introduction to JavaScript
Defining JavaScript: The Wizard of InteractivityJavaScript (JS) is a lightweight, fast programming language designed to make web pages interactive. It allows pages to respond to user actions rather than just displaying static information. If HTML is the structure and CSS is the style, JavaScript is the "brain" that makes elements move, change, and react.  



### 1.2. Why is JavaScript Essential?
Amazing Interactivity: It powers everything from clickable buttons to interactive photo galleries.  

Smart Data Validation: It verifies user input (like form data) before it is sent to the server, making processes faster and smoother.  

Full-Stack Development: Using Node.js, developers can use JavaScript for both front-end and back-end (server) development.  

Mobile Applications: Frameworks like React Native allow you to build mobile apps for iPhone and Android using JavaScript.  

Huge Community: A massive support network exists, providing endless libraries and resources for developers.


### 1.3. The Integrated Team: HTML, CSS, & JS
Think of these three as a team working together to build a website:

HTML (HyperText Markup Language): The skeleton. It defines the foundation (e.g., "There is a button here").  

CSS (Cascading Style Sheets): The appearance. It controls colors and fonts (e.g., "This button is blue and large").  

JavaScript: The mastermind. It controls the behavior (e.g., "When the user clicks the button, change the background color to yellow").




### 1.4. Practical Example

### **2. Operators: Your Tools for Changing Data**
Operators in JavaScript are the symbols we use to do specific things with data, like addition, subtraction, and comparison. Think of them like the tools in your toolbox, each tool has a specific function.

We use these operators to compare two values, and the result is always true or false :

### 2.4 Control Flow: Making Decisions in Code
Control flow allows your program to make decisions based on conditions (like: if something is true, do a specific action).

###### 2.4.1 if Statement (Basic Decision)

Used when you want to execute code only if a condition is true.

```javascript
const number = 12;

if (number > 10) {
  console.log("The number is greater than 10.");
}
```


###### 2.4.2 if...else (Two Choices)
Used when you have two possible outcomes.

```javascript
const isLoggedIn = false;

if (isLoggedIn) {
  console.log("Welcome!");
} else {
  console.log("Please log in.");
}
```

###### 2.4.3. if...else if...else (Multiple Choices)
Used when you have more than two conditions.
```javascript
const steps = 8000;

if (steps < 5000) {
  console.log("Walk more.");
} else if (steps < 10000) {
  console.log("Good job!");
} else {
  console.log("Excellent!");
}
```

### 3. Loops: Repeating Tasks

Loops are used to repeat code multiple times instead of writing it again and again.


---

3.1 for Loop (Known Number of Times)

Use when you know how many times the loop should run.
```javascript
for (let i = 1; i <= 5; i++) {
  console.log("Number: " + i);
}
```
Explanation:

- let i = 1 → start value
- i <= 5 → condition (loop runs while true)
- i++ → increase by 1 each time





3.2 while Loop (Unknown Number of Times)

Use when the loop depends on a condition.
```javascript
let i = 1;

while (i <= 5) {
  console.log("Counting: " + i);
  i++;
}
```
Explanation:
- Runs as long as the condition is true
- Must update the variable (i++) to avoid infinite loop





3.3. do...while Loop (Runs at Least Once)

Runs the code at least once, even if the condition is false.
```javascript
let i = 6;

do {
  console.log("Runs at least once");
  i++;
} while (i <= 5);
```
Explanation:
- Code runs first, then condition is checked

## 4. Functions – Organizing Code and Reusability

Functions are reusable blocks of code used to organize programs and avoid repetition.


---


4.1 Function Declaration (Basic Function)

Used to define and run a function.

```javascript
function sayHello() {
  console.log("Hello and welcome!");
}

sayHello();
```
4.2 Functions with Parameters

```javascript
Functions can take inputs (parameters).

function greetPerson(name) {
  console.log("Hello " + name + "!");
}

greetPerson("Maha");
greetPerson("Raneem");
```

4.3 Functions with Return Values

Used to return a result from a function.
```javascript
function addNumbers(a, b) {
  return a + b;
}

let result = addNumbers(5, 3);
console.log(result); // 8
```

```javascript
// Function expression
const multiply = function(a, b) {
  return a * b;
};

// Arrow function
const divide = (a, b) => a / b;
```

```javascript
function greet(name = "Guest") {
  console.log("Hello " + name + "!");
}

greet();
greet("Maha");
```

Difference:
- `console.log()` → prints output
- `return` → sends value for reuse

## 5. Collections (Objects and Arrays)

Collections are used to store and manage multiple related values in JavaScript.


---
5.1 Objects: Collections of Information

Objects store data as key-value pairs.
```javascript
let person = {
  name: "John",
  age: 30,
  isStudent: false
};
```
Accessing Properties
```javascript
// Dot notation
console.log(person.name);

// Bracket notation
console.log(person["age"]);
```
Adding / Updating Properties
```javascript
person.city = "New York";
person.age = 31;
```

Deleting Properties
```javascript
delete person.isStudent;
```
Nested Objects
```javascript
let student = {
  name: "Alice",
  contact: {
    email: "alice@example.com",
    phone: "123-456"
  }
};

console.log(student.contact.email);
```
5.2 Arrays: Ordered Lists

Arrays store multiple values in order (index starts from 0).
```javascript
let fruits = ["apple", "banana", "orange"];
```

Access & Modify
```javascript
console.log(fruits[0]);
fruits[1] = "mango";
```
Useful Array Methods
```javascript
fruits.push("grape");     // add end
fruits.unshift("kiwi");   // add start
fruits.pop();             // remove last
fruits.shift();           // remove first
```
Find & Check
```javascript
fruits.indexOf("apple");
fruits.includes("mango");
```

Looping
```javascript
for (let fruit of fruits) {
  console.log(fruit);
}
```
5.3 Arrays of Objects

Used to store multiple items with details.
```javascript
let tasks = [
  { id: 1, text: "Buy groceries", completed: false },
  { id: 2, text: "Clean house", completed: true }
];
```
Common Operations
```javascript
// Filter
tasks.filter(t => t.completed);

// Find
tasks.find(t => t.id === 1);

// Map
tasks.map(t => t.text);
```

Display Data
```javascript
tasks.forEach(task => {
  console.log(task.text);
});

```

## **8. Event Handling**
Events are the way JavaScript uses to respond to user actions. Like when the user clicks a button, types in a text field, or moves the mouse. JavaScript can "listen" to these events and react to them.


---


#### 8.1 Common Event Types

Most used events:
- `click` → when user clicks an element
- `mouseover` → when mouse hovers over an element
- `mouseout` → when mouse leaves an element
- `keydown` → when a key is pressed
- `submit` → when a form is submitted
- `load` → when the page finishes loading

#### 8.2 Adding Event Listeners

Used to detect and respond to events.
```javascript
let button = document.getElementById("my-button");

button.addEventListener("click", function() {
  alert("Button clicked!");
});
```
Using a separate function:
```javascript
function handleClick() {
  alert("Button clicked!");
}

button.addEventListener("click", handleClick);
```
#### 8.3 Event Object

Provides information about the event.
```javascript
button.addEventListener("click", function(event) {
  console.log(event.type);     // click
  console.log(event.target);   // clicked element
});
```

#### 8.4 Practical Examples
Change background color on click
```javascript
let btn = document.getElementById("color-btn");

btn.addEventListener("click", function() {
  document.body.style.backgroundColor = "lightblue";
});
```
Show / Hide element
```javascript
let toggleBtn = document.getElementById("toggle-btn");
let content = document.getElementById("content");

toggleBtn.addEventListener("click", function() {
  if (content.style.display === "none") {
    content.style.display = "block";
  } else {
    content.style.display = "none";
  }
});
```

## 9. Practical Project – Simple Calculator

Now it’s time to apply everything you learned in a small real project: a simple calculator

#### 9.1 Project Idea

You will build a calculator that can:

- Add two numbers
- Subtract two numbers
- Multiply two numbers
- Divide two numbers

#### 9.2 HTML Structure
```html
<h1>Simple Calculator</h1>

<input id="num1" type="number" placeholder="First number">
<input id="num2" type="number" placeholder="Second number">

<br><br>

<button id="addBtn">Add</button>
<button id="subBtn">Subtract</button>
<button id="mulBtn">Multiply</button>
<button id="divBtn">Divide</button>

<h2 id="result">Result: </h2>
```

#### 9.3 JavaScript Code
Get elements
```javascript
let num1 = document.getElementById("num1");
let num2 = document.getElementById("num2");
let result = document.getElementById("result");
```

Helper function to get values
```javascript
function getValues() {
  let a = Number(num1.value);
  let b = Number(num2.value);
  return [a, b];
}
```
Addition
```javascript
document.getElementById("addBtn").addEventListener("click", function() {
  let [a, b] = getValues();
  result.textContent = "Result: " + (a + b);
});
```
Subtraction
```javascript
document.getElementById("subBtn").addEventListener("click", function() {
  let [a, b] = getValues();
  result.textContent = "Result: " + (a - b);
});
```
Multiplication
```javascript
document.getElementById("mulBtn").addEventListener("click", function() {
  let [a, b] = getValues();
  result.textContent = "Result: " + (a * b);
});
```
Division (with error check)
```javascript
document.getElementById("divBtn").addEventListener("click", function() {
  let [a, b] = getValues();

  if (b === 0) {
    result.textContent = "Error: Cannot divide by 0";
  } else {
    result.textContent = "Result: " + (a / b);
  }
});
```
