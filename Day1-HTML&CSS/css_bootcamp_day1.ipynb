# **CSS (Cascading Style Sheets)**

---

## 1. What is CSS?

CSS stands for **Cascading Style Sheets**.

It is used to style and design web pages.

- HTML → structure
- CSS → appearance (colors, layout, fonts)

##### Without CSS, websites look plain.

## 2. Basic Syntax

CSS is written as:


```css
selector {
  property: value;
}
```

```css
p {
  color: red;
  font-size: 20px;
}
```

- selector → what element to style
- property → what to change
- value → how to change it

## 3. Selectors

Selectors are used to target HTML elements.
| Selector Type      | Syntax         | Description                               |
| ------------------ | ---------------------- | ----------------------------------------- |
| **Universal**      | `*`                    | All elements                              |
| **Type**           | `p`                    | All `<p>` elements                        |
| **Class**          | `.highlight`           | Elements with `class="highlight"`         |
| **ID**             | `#main`                | Element with `id="main"`                  |
| **Attribute**      | `input[type="text"]`   | Elements with specific attributes         |
| **Pseudo-class**   | `a:hover`              | Elements in a specific state              |


```css
/* Element selector */
p {
  color: blue;
}

/* Class selector */
.card {
  background: lightgray;
}

/* ID selector */
#title {
  font-size: 30px;
}
```

## 4. Combinators

Used to select elements based on relationships.

| Combinator        | Syntax   | Description                                         |
| ----------------- | -------- | ----------------------------------------------- |
| **Descendant**    | `div p`  | All `<p>` inside `<div>`                        |
| **Child**         | `div > p`| Direct `<p>` children of `<div>`                |
| **Adjacent**      | `h1 + p` | First `<p>` immediately after `<h1>`            |
| **General**       | `h1 ~ p` | All `<p>` after `<h1>` (same parent)             |

```css
/* Descendant (inside) */
div p {
  color: red;
}

/* Direct child */
div > p {
  color: blue;
}

/* Adjacent sibling */
h1 + p {
  color: green;
}
```


## 5. Variables and Root Scope

CSS variables help reuse values.
CSS custom properties (variables) are defined with --name and retrieved with var(--name). They are frequently defined in the :root pseudo-class for global application.

```css
:root {
  --main-color: blue;
  --padding: 10px;
}

.box {
  color: var(--main-color);
  padding: var(--padding);
}
```

## 6. Sizing

Used to control width and height.
*   width, height → set exact size (px, em, %, vw, vh).

*   fit-content → grows just enough to fit the content, but can wrap.

*   min-width, max-width → set limits so it can’t shrink or grow beyond these.

*   auto → size is decided by the content or parent.


| Unit  | Explaination                      |
| ----- | ---------------------------- |
| `px`  | Pixels                       |
| `%`   | Percentage relative to the parent|
| `em`  | Relative to parent font size |
| `rem` | Relative to root font size   |
| `vh`  | % of viewport height         |
| `vw`  | % of viewport width          |

```css
.box {
  width: 200px;
  height: 100px;
}
```

## 7. Box Model

Every HTML element behaves like a **box** made of four layers:

1. **Content** → the actual text or image  
2. **Padding** → space inside the element (around content)  
3. **Border** → the outline around the element  
4. **Margin** → space outside the element  

---

### 🔹 Margin (Outer Space)

Margin controls the space **outside** the element.

| Property | Description |
|----------|------------|
| `margin: 10px;` | All sides = 10px |
| `margin: 10px 20px;` | Top/bottom = 10px, left/right = 20px |
| `margin: 10px 20px 30px;` | Top = 10px, sides = 20px, bottom = 30px |
| `margin: 10px 20px 30px 40px;` | Top, right, bottom, left (clockwise) |
| `margin-top`, `margin-right`, etc. | Control each side |
| `margin: auto;` | Centers element horizontally (with width) |

---

### 🔹 Padding (Inner Space)

Padding controls the space **inside** the element.

✔ Same shorthand rules as margin

---

### 🔹 Border

Border wraps around content and padding.

| Property | Description | Example |
|----------|------------|--------|
| `border` | shorthand (width, style, color) | `2px solid red` |
| `border-width` | thickness | `2px` |
| `border-style` | style | `solid` |
| `border-color` | color | `red` |
| `border-radius` | rounded corners | `12px` |

---
### code :
```html
<!DOCTYPE html>
<html>
<head>
  <style>
    .card {
      width: 300px;
      height: 200px;
      padding: 20px;
      border: 2px solid black;
      margin: 20px;
    }
  </style>
</head>
<body>
  <div class="card">This is a card.</div>
</body>
</html>
```

### 🔹 Box Sizing

Controls how width and height are calculated:

- `content-box` → width = content only  
- `border-box` → width = content + padding + border  

---
```css
<style>
  .cb {
    box-sizing: content-box;
    width: 200px;
    padding: 20px;
    border: 5px solid blue;
  }

  .bb {
    box-sizing: border-box;
    width: 200px;
    padding: 20px;
    border: 5px solid green;
  }
</style>

<div class="cb">Content-box</div>
<div class="bb">Border-box</div>
```

### 🌫️ Shadows

Used to add visual depth.

| Property | Example |
|----------|--------|
| `box-shadow` | `0 0 10px gray;` |
| `text-shadow` | `0 0 10px black;` |

---

### 🔹 Overflow

Controls what happens when content is larger than its container.

| Value | Behavior |
|------|--------|
| `visible` | content overflows (default) |
| `hidden` | extra content is clipped |
| `scroll` | always shows scrollbars |
| `auto` | scrollbars appear only if needed |

---

```css
.container {
  width: 200px;
  height: 100px;
  overflow: auto;
}
```

## 8. Typography

Controls text appearance.
| Property          | Example                               | Description                   |
| ----------------- | ------------------------------------- | ----------------------------- |
| `font-family`     | `'Arial', sans-serif`                 | Font type                     |
| `font-size`       | `16px`                                | Font size                     |
| `color`           | `#333`                                | Text color                    |
| `line-height`     | `1.5`                                 | Space between lines           |
| `text-align`      | `center`                              | Text alignment                |
| `text-transform`  | `uppercase`                           | Case transformation           |
| `text-decoration` | `underline`                           | Underline, line-through, etc. |

```css
body {
  font-family: Arial;
  font-size: 16px;
  color: #333;
}

h1 {
  font-weight: bold;
}
```

## 9. Display & Positioning

**Used to control layout and visibility.**

| Display Value     | Description                   |
| ----------------- | ------------------------- |
| `block`           | New line, full width      |
| `inline`          | Flows inline with text    |
| `inline-block`    | Inline + block features   |
| `flex`            | Flexbox layout            |
| `grid`            | Grid layout               |
| `none`            | Hides element             |

| Position Value | Description                                         |
| -------------- | --------------------------------------------------- |
| `static`       | Default; follows normal flow                        |
| `relative`     | Moves relative to normal position                   |
| `absolute`     | Positioned to nearest positioned ancestor           |
| `fixed`        | Stays fixed to viewport                             |
| `sticky`       | Switches between relative and fixed based on scroll |


`z-index` controls stack order.
##_______________________________________
### code (display) :
```css
div {
  display: block;
}

span {
  display: inline;
}
```
##_______________________________________
### code (Flexbox) :
```css
.container {
  display: flex;
  justify-content: center;
  align-items: center;
}
```

```css
.parent {
  position: relative;
}

.child {
  position: absolute;
  top: 0;
  right: 0;
}
```

## 11. Transitions & Animations

Used to add **movement and interactivity** to elements.

- **Transition** → smooth change between states (e.g., hover)
- **Animation** → more complex movement using keyframes

---

### 🔹 Transition

Allows properties to change smoothly instead of instantly.

```css
transition: property duration timing-function delay;
```
Example :
```css
transition: all 0.3s ease-in-out;
```

Parts:

* **property** → what to animate (`color`, `background`)
* **duration** → how long (`0.5s`)
* **timing-function** → speed curve (`ease`, `linear`)
* **delay** → wait time before starting


**`transform`**

Used to rotate, scale, move elements.

```css
transform: rotate(45deg) translate(10px, 20px);
```

| Function      | Description                  |
| ------------- | ---------------------------- |
| `rotate()`    | Rotates element              |
| `translate()` | Moves element along X/Y axis |
| `scale()`     | Scales size                  |
| `skew()`      | Skews element                |

```html
<!DOCTYPE html>
<html>
<head>
  <style>
    /* Transition example */
    .button {
      background: lightblue;
      transition: background 0.3s, color 0.3s;
    }

    .button:hover {
      background: navy;
      color: white;
    }

    /* Animation example */
    @keyframes slideIn {
      from {
        transform: translateX(-100%);
      }
      to {
        transform: translateX(0);
      }
    }

    .animated {
      animation: slideIn 1s ease-out forwards;
    }
  </style>
</head>
<body>

  <button class="button">Hover me</button>

  <div class="animated">Sliding in!</div>

</body>
</html>
```

### 🔹 Animation

Animations use `@keyframes` to define steps of movement.

- `from` → starting state  
- `to` → ending state  

You can also use percentages (0% → 100%).

---

### Animation Properties

| Property | Description |
|----------|------------|
| `animation-name` | name of keyframes |
| `animation-duration` | how long it runs |
| `animation-timing-function` | speed style |
| `animation-fill-mode` | keeps final state |
| `animation-iteration-count` | number of repeats |

---

### 💡 Quick Tip

- Use **transition** for simple effects (hover)
- Use **animation** for complex movement

## 12. Media Queries (Responsive Design)

Used to make websites work on different screen sizes.
```css
@media (max-width: 600px) {
  body {
    background: lightgray;
  }
}
```
