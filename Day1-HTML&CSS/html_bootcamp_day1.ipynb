# What is HTML?

**HTML (HyperText Markup Language)** is the standard language used to create webpages.

HTML describes the **structure of a webpage** using elements called **tags**.

Example:

```html
<h1>Hello World</h1>
<p>This is my first webpage.</p>
```

The browser reads this code and displays it visually.

 **Note** HTML files end with `.html`

---

# Basic HTML Document Structure

Every HTML page follows this basic structure:

```html
<!DOCTYPE html>
<html lang="en">

<head>
    <meta charset="UTF-8">
    <title>My First Page</title>
</head>

<body>

    <h1>Hello World</h1>
    <p>Welcome to web development!</p>

</body>

</html>
```

### Important Parts

**`<!DOCTYPE html>`**  
Tells the browser that this is an HTML5 document.

**`<html>`**  
The root element of the webpage.

**`<head>`**  
Contains metadata like:
- page title
- character encoding
- links to CSS

**`<body>`**  
Contains everything that appears on the webpage.

# HTML Attributes

**HTML attributes** provide **additional information** about HTML elements. They are always specified in the **opening tag** and usually come in **name="value"** pairs.

---
**Basic Syntax**

```html
<tagname attribute1="value1" attattribute2="value2" attribute3>Content</tagname>
```

Example:

```html
<a href="https://example.com">Visit</a>
```

Here:

* `a` is the tag
* `href` is the attribute
* `"https://example.com"` is the attribute's value

---

## Types of HTML Attributes

HTML attributes can be categorized into **two main types**:


 **1. String Attributes**

These attributes require a **value** enclosed in **quotes** (`" "` or `' '`), usually describing **URLs, text, names, CSS styles**, etc.

##### Examples:

```html
<a href="https://example.com">Link</a>       <!-- href = string -->
<img src="image.png" alt="A sample image">  <!-- src and alt = string -->
<input type="text" placeholder="Enter name"> <!-- type and placeholder = string -->
```

 **2. Boolean Attributes**

These are either **present or absent**. If the attribute is present, it's considered **true**. If missing, it's **false**—they don’t need a value.

##### Examples:

```html
<input type="checkbox" checked>
<input type="text" disabled>
<button autofocus>Click Me</button>
```

In these examples:

* `checked`, `disabled`, and `autofocus` are **boolean attributes**.
* Writing `checked="checked"` is also valid but not required.


#### Summary Table

| Type    | Needs Value? | Example                | Description                 |
| ------- | ------------ | ---------------------- | --------------------------- |
| String  | ✅ Yes        | `<img src="logo.png">` | Source is a string URL      |
| Boolean | ❌ No         | `<input disabled>`     | Disables input when present |

# Layout & Structure Tags
---

### `<main>`

* **Description**: The main content unique to the page.
* **Tip**: There should be only one `<main>` per page.


```html
<main>
  <article>
    <h2>Blog Post</h2>
    <p>Content goes here...</p>
  </article>
</main>
```
Output :

<main>
  <article>
    <h2>Blog Post</h2>
    <p>Content goes here...</p>
  </article>
</main>

---

### `<header>`

* **Description**: Headings are used for titles and section headers.

```html
<h1>Main Title</h1>
<h2>Section Title</h2>
<h3>Subsection</h3>
```
Output :

<h1>Main Title</h1>
<h2>Section Title</h2>
<h3>Subsection</h3>

- `h1` is the most important heading.
- `h6` is the smallest.

---

### `<Paragraphs>`

* **Description**: Paragraphs are used for text content.

```html
<p>This is a paragraph.</p>
```
Output :

<p>This is a paragraph.</p>

---

### `<nav>`

* **Description**: Defines a section for navigation links.
* **Used for**: Menus, sidebars, table of contents.

```html
<nav>
  <a href="#home">Home</a>
  <a href="#about">About</a>
</nav>
```
Output :

<nav>
  <a href="#home">Home</a>
  <a href="#about">About</a>
</nav>

---

### `<section>`

* **Description**: A standalone, thematically related group of content.
* **Best used for**: Dividing the page into logical sections.

```html
<section>
  <h2>Services</h2>
  <p>We offer...</p>
</section>
```
Output :

<section>
  <h2>Services</h2>
  <p>We offer...</p>
</section>

---

### `<div>`

* **Description**: The `div` element is used to group elements together.

```html
<div>
    <h2>About Me</h2>
    <p>I am learning web development.</p>
</div>
```
Output :

<div>
    <h2>About Me</h2>
    <p>I am learning web development.</p>
</div>

---

### `<footer>`

* **Description**: Represents the bottom section—copyrights, contact links.


```html
<footer>
  <p>&copy; 2025 My Site</p>
</footer>
```
Output :

<footer>
  <p>&copy; 2025 My Site</p>
</footer>

# Links 

---


Links allow users to navigate between pages. using the **`<a>`** tag.

* **Common Attributes**:

  * `href`: The URL the link points to.
  * `target`: Where to open the link (`_blank` = new tab).

```html
<a href="https://google.com">Visit Google</a>
```

Open link in a new tab:

```html
<a href="https://google.com" target="_blank">Open Google</a>
```

# Lists

---
### Unordered List

* **Description**:  The  `<ul>` element is used to create a bulleted list of items. Unlike orderd list, the order of the items does not matter.

```html
<ul>
  <li>HTML</li>
  <li>CSS</li>
  <li>JavaScript</li>
</ul>
```
Output :

<ul>
  <li>HTML</li>
  <li>CSS</li>
  <li>JavaScript</li>
</ul>

---
### Ordered List

* **Description**: `<ol>`element is used to create a list of items in a specific order, typically numbered.

* **Attributes**: type, start, reversed

```html
<ol type="A" start="3" reversed>
  
  <li>Third Item</li>
  <li>Second Item</li>
  <li>First Item</li>

</ol>
```
• type="A" → Uses uppercase letters (A, B, C…)

• start="3" → Starts from C (3rd letter)

• reversed → Reverses order → C, B, A


Output :

<ol type="A" start="3" reversed>
  
  <li>Third Item</li>
  <li>Second Item</li>
  <li>First Item</li>

</ol>

# Form Elements

---

### `<form>`

* **Description**: Creates a container for user input elements to collect and send data.
* **Common Attributes**:

  * `action`, `method`

  •	action: The URL where the form data is sent for processing.

  • method: HTTP method used to send data (GET or POST).

---

### `<input>`


* **Description**: Defines various input fields inside a form.

* **Common Attributes**:
  * `type`, `name`, `value`, `placeholder`,`required`, `id`,

 **• type:** Type of input (e.g., `text`, `email`, `password`, `checkbox`, `radio`, `submit`).

 **• name:** Name of the input used as the key when submitting data.

 **• value:** The initial or submitted value of the input.

 **• placeholder:** Hint text shown inside the input.

 **• required:** Marks the input as mandatory.

 **• id:** Unique identifier to link with a label
  
---

###  `<label>`


* **Description**: Associates text with a specific input element, improving accessibility.
* **Common Attributes**:

	`for`: The id of the input element it labels.
  
---

###  `<textarea>`


* **Description**: Defines a multi-line text input.
* **Common Attributes**:
  * `name`, `rows`, `cols`,`placeholder`

 **• name:** Name of the textarea.

 **• rows:** Number of visible text lines.

 **• cols:** Visible width in characters.

 **• placeholder:** Hint text inside the textarea.
  
---


###  `<button>`


* **Description**: Defines a clickable button.
* **Common Attributes**:

 **• type:** Button behavior (`submit`, `reset`, `button`).
  
---
#### Example:
```html
<form action="/submit-form" method="POST">
  <label for="name">Name:</label><br>
  <input type="text" id="name" name="name" placeholder="Your name" required><br><br>

  <label for="email">Email:</label><br>
  <input type="email" id="email" name="email" placeholder="you@example.com" required><br><br>

  <label for="message">Message:</label><br>
  <textarea id="message" name="message" rows="4" cols="40" placeholder="Your message"></textarea><br><br>

  <button type="submit">Submit</button>
</form>
```

# Basic Page Example

```html
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <title>My Portfolio</title>
</head>
<body>

<!-- Header -->
<header>
    <h1>Raneem Alhajri Portfolio</h1>
    <nav>
        <a href="#about">About Me</a> |
        <a href="#skills">Skills</a> |
        <a href="#projects">Projects</a> |
        <a href="#contact">Contact</a>
    </nav>
</header>

<!-- Main Content -->
<main>

    <!-- About Section -->
    <section id="about">
        <h2>About Me</h2>
        <p>Hello! I am Raneem, a Cybersecurity and Web Development enthusiast. I love learning and building projects.</p>
        <img src="profile.jpg" alt="Profile Picture" width="150">
    </section>

    <!-- Skills Section -->
    <section id="skills">
        <h2>My Skills</h2>
        <ul>
            <li>HTML & CSS</li>
            <li>JavaScript</li>
            <li>Python</li>
            <li>Cybersecurity</li>
        </ul>
    </section>

    <!-- Projects Section -->
    <section id="projects">
        <h2>Projects</h2>
        <ol type="A">
            <li>Personal Portfolio Website</li>
            <li>Mini Blog Platform</li>
            <li>Online Library System</li>
        </ol>
    </section>

    <!-- Contact Form -->
    <section id="contact">
        <h2>Contact Me</h2>
        <form>
            <input type="text" placeholder="Your Name" required>
            <input type="email" placeholder="Your Email" required>
            <textarea placeholder="Your Message"></textarea>
            <button type="submit">Send</button>
        </form>
    </section>

</main>

<!-- Footer -->
<footer>
    <p>&copy; 2026 Raneem Alhajri. All Rights Reserved.</p>
    <a href="https://linkedin.com" target="_blank">LinkedIn</a> |
    <a href="https://github.com" target="_blank">GitHub</a>
</footer>

</body>
</html>
```

<th>Age</th>
</tr>

<tr>
<td>Ali</td>
