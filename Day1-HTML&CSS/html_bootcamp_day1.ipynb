# What is HTML?

**HTML (HyperText Markup Language)** is the standard language used to create webpages.

HTML describes the **structure of a webpage** using elements called **tags**.

Example:

```html
<h1>Hello World</h1>
<p>This is my first webpage.</p>
```

The browser reads this code and displays it visually.

---

# Basic HTML Document Structure

Every HTML page follows this basic structure:

```html
<!DOCTYPE html>
<html lang="en">

<head>
    <meta charset="UTF-8">
    <title>My First Page</title>
</head>

<body>

    <h1>Hello World</h1>
    <p>Welcome to web development!</p>

</body>

</html>
```

### Important Parts

**`<!DOCTYPE html>`**  
Tells the browser that this is an HTML5 document.

**`<html>`**  
The root element of the webpage.

**`<head>`**  
Contains metadata like:
- page title
- character encoding
- links to CSS

**`<body>`**  
Contains everything that appears on the webpage.

---

# Headings

Headings are used for titles and section headers.

```html
<h1>Main Title</h1>
<h2>Section Title</h2>
<h3>Subsection</h3>
```

- `h1` is the most important heading.
- `h6` is the smallest.

---

# Paragraphs

Paragraphs are used for text content.

```html
<p>This is a paragraph.</p>
```

---

# Links

Links allow users to navigate between pages.

```html
<a href="https://google.com">Visit Google</a>
```

Open link in a new tab:

```html
<a href="https://google.com" target="_blank">Open Google</a>
```

---

# Images

Images are added using the `img` tag.

```html
<img src="image.jpg" alt="Description of image">
```

Important attributes:

- `src` → image path
- `alt` → description for accessibility

---

# Lists

## Unordered List

```html
<ul>
  <li>HTML</li>
  <li>CSS</li>
  <li>JavaScript</li>
</ul>
```

## Ordered List

```html
<ol>
  <li>Step One</li>
  <li>Step Two</li>
</ol>
```

---

# Div (Container)

The `div` element is used to group elements together.

```html
<div>
    <h2>About Me</h2>
    <p>I am learning web development.</p>
</div>
```

---

# Basic Page Example

```html
<!DOCTYPE html>
<html>

<head>
    <title>My Website</title>
</head>

<body>

    <h1>My Website</h1>

    <p>Welcome to my first webpage.</p>

    <h2>My Skills</h2>

    <ul>
        <li>HTML</li>
        <li>CSS</li>
        <li>JavaScript</li>
    </ul>

</body>

</html>
```

## Semantic HTML

Examples:

```html
<header>
<nav>
<main>
<section>
<footer>
```

Semantic elements help make webpages:
- easier to understand
- better for accessibility
- better for SEO

---

## Forms

Forms allow users to send data.

```html
<form>
    <input type="text" placeholder="Enter your name">
    <button>Submit</button>
</form>
```

Used for:
- login pages
- sign up pages
- contact forms

---

## Tables

Tables display data in rows and columns.

```html
<table>
<tr>
<th>Name</th>
<th>Age</th>
</tr>

<tr>
<td>Ali</td>
<td>21</td>
</tr>

</table>
```
