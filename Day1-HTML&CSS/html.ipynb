## 📌 How Websites Work

A website has 3 main parts:

1. **HTML** → Structure
2. **CSS** → Style
3. **JavaScript** → Interactivity

Think of it like a house:
- HTML = walls
- CSS = design
- JavaScript = behavior

---

## 📌 What is HTML?

HTML = **HyperText Markup Language**.

It is used to **structure content** on webpages.

- Not a programming language ❌
- It's a markup language ✅

Think of HTML like **Microsoft Word but using code**.

HTML files end with `.html`


## 🧱 Basic HTML Structure

```html
<!DOCTYPE html>
<html>
<head>
    <title>My First Website</title>
</head>
<body>

    <h1>Hello Bootcamp!</h1>
    <p>This is my first website.</p>

</body>
</html>
```

### Explanation:
- `<html>` → whole page
- `<head>` → info about page
- `<body>` → visible content
- `<h1>` → heading
- `<p>` → paragraph


## ⭐ Important HTML Tags

| Tag | Purpose |
|-----|--------|
| `<h1> - <h6>` | Headings |
| `<p>` | Paragraph |
| `<a>` | Link |
| `<img>` | Image |
| `<ul> <li>` | List |
| `<div>` | Container |

---

### Example:
```html
<h1>Title</h1>
<p>This is text</p>
```


## 🔗 Links

```html
<a href="https://google.com">Visit Google</a>
```

Open in new tab:

```html
<a href="https://google.com" target="_blank">Open Google</a>
```


## 🖼 Images

```html
<img src="image.jpg" alt="Description">
```

- `src` → image path
- `alt` → description


## 📋 Lists

### Unordered List
```html
<ul>
  <li>HTML</li>
  <li>CSS</li>
</ul>
```

### Ordered List
```html
<ol>
  <li>Step 1</li>
  <li>Step 2</li>
</ol>
```


## 📦 Div (Container)

```html
<div>
    <h2>About Me</h2>
    <p>I love coding</p>
</div>
```

Used to group elements.


## 🧠 HTML Syntax

Basic format:

```html
<tag>Content</tag>
```

- Opening tag
- Content
- Closing tag

---

## ⚠️ Self Closing Tags

```html
<img />
<br />
<hr />
```


## ⚙️ Attributes

Attributes give extra info to tags.

```html
<a href="https://example.com">Link</a>
```

### Types:

### 1. String
```html
<img src="img.png">
```

### 2. Boolean
```html
<input disabled>
```


## 💻 Full Example

```html
<!DOCTYPE html>
<html>
<head>
    <title>My Website</title>
</head>
<body>

<h1>My Website</h1>
<p>Hello!</p>

<a href="https://google.com">Google</a>

<ul>
    <li>HTML</li>
    <li>CSS</li>
</ul>

</body>
</html>
```
