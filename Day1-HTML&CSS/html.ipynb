## üìå What is HTML?

HTML = HyperText Markup Language.

###### **HTML** stands for **HyperText Markup Language**. It is the standard language used to create and structure content on the web. HTML is not a programming language‚Äîit‚Äôs a markup language, which means it uses tags to label elements like headings, paragraphs, links, images, tables, and more. **You can think of HTML as Microsoft Word but with using codes**.


HTML files end with `.html`

## First HTML Code
```html
<!DOCTYPE html>
<html>
<head>
    <title>My First Website</title>
</head>
<body>

<h1>Hello Bootcamp!</h1>
<p>This is my first website.</p>

</body>
</html>
```

Explain:
```
- <h1> heading
- <p> paragraph
- <body> content
```

### **2.a. HTML Tags**
HTML tags are special keywords used to define elements on a web page.  
They organize content such as text, images, links, and sections so the browser knows how to display them.


---

#### **2.a.i. Syntax of an HTML Tag**

```html
<tagname>Content</tagname>
```

* Most HTML tags come in **pairs**:

  * **Opening tag**: `<tagname>`
  * **Closing tag**: `</tagname>`
  * **Content**: the part between the tags; can be text or **HTML Tags**.

Example:

```html
<p>This is a paragraph.</p>
<div>
    <article>
        <h1>Hi</h1>
        <section>
            <p>HTMl is the best</p>
        </section>
    </article>
<div>
```
---

##### **2.a.ii. Self-Closing Tags**

Some tags **don‚Äôt have content** and are **self-closing**, like:

```html
<img src="image.jpg" alt="My Image" />
<br />
<hr />
```
---

#### **2.b. HTML Attributes**

**HTML attributes** provide **additional information** about HTML elements. They are always specified in the **opening tag** and usually come in **name="value"** pairs.

---
##### **2.b.i. Basic Syntax**

```html
<tagname attribute1="value1" attattribute2="value2" attribute3>Content</tagname>
```

Example:

```html
<a href="https://example.com">Visit</a>
```

Here:

* `a` is the tag
* `href` is the attribute
* `"https://example.com"` is the attribute's value

---

### **2.b.ii. Types of HTML Attributes**

HTML attributes can be categorized into **two main types**:


#### **1. String Attributes**

These attributes require a **value** enclosed in **quotes** (`" "` or `' '`), usually describing **URLs, text, names, CSS styles**, etc.

##### Examples:

```html
<a href="https://example.com">Link</a>       <!-- href = string -->
<img src="image.png" alt="A sample image">  <!-- src and alt = string -->
<input type="text" placeholder="Enter name"> <!-- type and placeholder = string -->
```

#### **2. Boolean Attributes**

These are either **present or absent**. If the attribute is present, it's considered **true**. If missing, it's **false**‚Äîthey don‚Äôt need a value.

##### Examples:

```html
<input type="checkbox" checked>
<input type="text" disabled>
<button autofocus>Click Me</button>
```

In these examples:

* `checked`, `disabled`, and `autofocus` are **boolean attributes**.
* Writing `checked="checked"` is also valid but not required.
