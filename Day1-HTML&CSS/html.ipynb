## 📌 What is HTML?

HTML = **HyperText Markup Language**

- Not a programming language ❌
- It's a markup language ✅

Used to structure web pages.


## 🧱 Basic Structure

```html
<!DOCTYPE html>
<html>
<head>
  <title>My Page</title>
</head>
<body>
  <h1>Hello</h1>
</body>
</html>
```


## ⭐ Common Tags

- `<h1>` Heading
- `<p>` Paragraph
- `<a>` Link
- `<img>` Image
- `<ul>` List
- `<div>` Container


## 🔗 Links

```html
<a href="https://google.com">Google</a>
```


## 🖼 Images

```html
<img src="image.jpg" alt="desc">
```


## 📋 Lists

```html
<ul>
  <li>Item</li>
</ul>
```
